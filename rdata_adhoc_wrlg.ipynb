## 1. Getting Started

- Importing libraries and connect to the db
- Structuring data for loading correctly

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('database/raw.db')
```

```python
tables = conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
tables = [table[0] for table in tables]

dfs = {'gram_table': [], 'gram_chart': [], 'tktk_table': [], 'tktk_chart': []}

for table in tables:
    for key in dfs.keys():
        if f'raw_{key}' in table:
            dfs[key].append(pd.read_sql_query(f'SELECT * FROM {table}', conn))

gram_table_df = pd.concat(dfs['gram_table'], ignore_index=True)
gram_chart_df = pd.concat(dfs['gram_chart'], ignore_index=True)
tktk_table_df = pd.concat(dfs['tktk_table'], ignore_index=True)
tktk_chart_df = pd.concat(dfs['tktk_chart'], ignore_index=True)
```
